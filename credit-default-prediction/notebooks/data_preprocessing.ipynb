```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# display aesthetics
from IPython.display import display
%matplotlib inline
pd.set_option('display.max_columns', 55)
pd.options.display.float_format = '{:.5f}'.format

import warnings
warnings.filterwarnings('ignore')
```

```python
# load the dataset
data = pd.read_csv('data.csv')
data.head()
```

```python
# drop variables with more than 50% missing values

cols = ['form_field40','form_field31','form_field41','form_field45','form_field11','form_field23']

data.drop(columns=cols, axis=1, inplace=True)
data.head()
```

```python
# Handling Categorical variables
data = pd.get_dummies(data, columns = ['form_field47'])
data.head()
```

```python
# drop applicant ID column

data.drop(columns = 'Applicant_ID', inplace=True, axis=1)
```

```python
data.columns
```

```python
# impute missing values using mean

miss_cols = ['form_field1', 'form_field2', 'form_field3', 'form_field4',
       'form_field5', 'form_field6', 'form_field7', 'form_field8',
       'form_field9', 'form_field10', 'form_field12', 'form_field13',
       'form_field14', 'form_field15', 'form_field16', 'form_field17',
       'form_field18', 'form_field19', 'form_field20', 'form_field21',
       'form_field22', 'form_field24', 'form_field25', 'form_field26',
       'form_field27', 'form_field28', 'form_field29', 'form_field30',
       'form_field32', 'form_field33', 'form_field34', 'form_field35',
       'form_field36', 'form_field37', 'form_field38', 'form_field39',
       'form_field42', 'form_field43', 'form_field44', 'form_field46',
       'form_field48', 'form_field49', 'form_field50']

for col in miss_cols:
    data[col] = data[col].fillna(data[col].mean())
    
data.isnull().sum()
```

```python
data.describe()
```

There is need to normalize the data so that every variable lies between 0 and 1. In this case, I will use MinMaxScaler from scikit-learn.

```python
# separate the datasets

train_set = data[data['set'] == 'train']
test_set = data[data['set'] == 'test']

test_set.drop(columns = ['default_status', 'set'], axis=1, inplace = True)
train_set.drop(columns = ['set'], axis=1, inplace=True)

X = train_set.drop(columns='default_status', axis=1)
y = train_set['default_status']
```

```python
# data normalization

from sklearn.preprocessing import MinMaxScaler

normalizer = MinMaxScaler()

train_set = normalizer.fit_transform(X)
test_set = normalizer.fit_transform(test_set)
```

```python
train_set = pd.DataFrame(train_set)
train_set.describe()
```

```python
test_set = pd.DataFrame(test_set)
test_set.describe()
```

```python
#Setting up training and validation sets

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression

X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
#Print shape for each set
print('X_train: ', X_train.shape)
print('X_val: ', X_val.shape)
print('y_train: ', y_train.shape)
print('y_val: ', y_val.shape)
```

```python
# save the datasets

X_train.to_csv('X_train.csv', index=False)
X_val.to_csv('X_val.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_val.to_csv('y_val.csv', index=False)
```
